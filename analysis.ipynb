```python
hist_counts = {}
for y in range(2000,2021):
    q = f'''
    SELECT y{y}, COUNT(y{y})
    FROM history as h 
    WHERE h.taxonid IN (
    SELECT DISTINCT(hab.taxonid)
    FROM habitat as hab
    WHERE hab.code IN ('1.5', '1.6', '1.7', '1.8', '1.9', '14.6'))
    GROUP BY y{y}
    '''
    curs = conn.cursor()
    curs.execute(q)
    hist_counts[y] = curs.fetchall()
hist_counts
columns = ['year', 'category', 'count']
lists = []
for i in hist_counts:
    for c in hist_counts[i]:
        lists.append([i,c[0], c[1]])
hist_counts = pd.DataFrame(lists, columns=columns)
hist_counts.to_csv('historicalCategoryCounts', index=False)
```

```python
class_hist = {}
for y in range(2000,2021):
    q = f'''
    SELECT a.class_name, h.y{y}, COUNT(*)
    FROM history as h JOIN assessment as a ON(h.taxonid=a.taxonid)
    WHERE h.taxonid IN (
    SELECT DISTINCT(hab.taxonid)
    FROM habitat as hab
    WHERE hab.code IN ('1.5', '1.6', '1.7', '1.8', '1.9', '14.6'))
    GROUP BY a.class_name, h.y{y}
    ORDER BY a.class_name
    '''
    curs = conn.cursor()
    curs.execute(q)
    class_hist[y] = curs.fetchall()
class_hist
columns = ['year', 'class', 'category', 'count']
lists = []
for i in class_hist:
    for c in class_hist[i]:
        lists.append([i,c[0], c[1], c[2]])
class_hist = pd.DataFrame(lists, columns=columns)
class_hist.to_csv('historicalCountsByClass', index=False)
```
