```python
import json
import pandas as pd
import numpy as np
```

```python
 def to_df(c):
  with open(f'data/all{c}species.json', 'r') as f:
    data = json.load(f)
    df = pd.DataFrame(data['result'])
  return df

cameroon = to_df('CM')
gabon = to_df('GA')
congo = to_df('CG')
drc = to_df('CD')
eg = to_df('GQ')
car = to_df('CF')
```

```python
all_species = cameroon.copy()
for df in [gabon, congo, drc, eg, car]:
    all_species = all_species.append(df)
    
all_species = all_species.drop_duplicates(['taxonid'])
all_species.drop('country', axis=1, inplace=True)
```

```python
x = {'taxonid':[],'code':[], 'habitat':[], 'suitability':[], 'season':[], 'majorimportance':[]}
with open('data/sp_habitats.json', 'r') as f:
    data = json.load(f)
for i in data:
    for r in i['result']:
        x['taxonid'].append(i['id'])
        x['code'].append(r['code'])
        x['habitat'].append(r['habitat'])
        x['suitability'].append(r['suitability'])
        x['season'].append(r['season'])
        x['majorimportance'].append(r['majorimportance'])
            
habitats = pd.DataFrame(x)
habitats.head(20)
```

```python
x = {'taxonid':[],'code':[], 'country':[], 'presence':[], 'origin':[], 'distribution':[]}
with open('data/sp_ctry_occur.json', 'r') as f:
    data = json.load(f)

for i in data:
    for r in i['result']:
        x['taxonid'].append(i['name'])
        x['code'].append(r['code'])
        x['country'].append(r['country'])
        x['presence'].append(r['presence'])
        x['origin'].append(r['origin'])
        x['distribution'].append(r['distribution_code'])
country_occurence = pd.DataFrame(x)
country_occurence.head()
```

```python
x = {'taxonid':[]}
for year in range(1988, 2021):
    x[year] = []
    
with open('data/sp_hist_cats.json', 'r') as f:
    data = json.load(f)

for i in data:
    change_yrs = {}
    ind = 0
    start = int(i['result'][ind]['year'])
    while start > 1988:
        try:
            next = i['result'][ind+1]['year']
            change_yrs[start] = i['result'][ind]['category']
            ind += 1
            start = int(next)
        except:
            break
    x['taxonid'].append(i['name'])
    curr_cat = i['result'][ind]['category']
    
    for year in range(1982, 2021):
        if year < start:
            x[year].append(np.nan)
        else:
            if year in change_yrs:
                curr_cat = change_yrs[year]
            x[year].append(curr_cat)

pd.DataFrame(x)
```
